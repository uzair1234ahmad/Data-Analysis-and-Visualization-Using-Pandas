```python
# Data loading
```

```python
import pandas as pd

# Load the Titanic dataset
titanic_df = pd.read_csv('train.csv')

# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Data Exploration
```

```python
# Display the structure of the dataset
titanic_df.info()

# Display summary statistics of the dataset
titanic_df.describe()

# Display the first few rows of the dataset to understand its features
titanic_df.head()
```

```python
# Data Cleaning
```

```python
# Check for missing values
missing_values = titanic_df.isnull().sum()

# Handle missing values
# For simplicity, let's fill missing 'Age' values with the median age and drop rows with missing 'Embarked' values
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df.dropna(subset=['Embarked'], inplace=True)

# Drop the 'Cabin' column due to a large number of missing values
titanic_df.drop(columns=['Cabin'], inplace=True)

# Check for duplicates and drop them if any
titanic_df.drop_duplicates(inplace=True)

# Verify that data cleaning was successful
titanic_df.info()
```

```python
# Data visulaization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Survival Rate by Class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

# Visualization 2: Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Visualization 3: Survival Rate by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

# Visualization 4: Fare Distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

Analysis and Insights
Survival Rate by Class:

Visualization: Bar plot of survival rate by passenger class.
Analysis: Passengers in the first class had a higher survival rate compared to those in the second and third classes.
Insights: Higher social and economic status might have contributed to a higher chance of survival.
Age Distribution:

Visualization: Histogram of age distribution with a KDE overlay.
Analysis: The age distribution shows that the majority of passengers were between 20 and 40 years old.
Insights: Understanding the age distribution can help in analyzing survival rates across different age groups.
Survival Rate by Gender:

Visualization: Bar plot of survival rate by gender.
Analysis: Females had a significantly higher survival rate compared to males.
Insights: This could be due to the "women and children first" policy during the evacuation.
Fare Distribution:

Visualization: Histogram of fare distribution with a KDE overlay.
Analysis: The fare distribution is right-skewed, with most passengers paying lower fares, but a few paying very high fares.
Insights: The wide range of fares indicates a diverse socioeconomic background of the passengers.
